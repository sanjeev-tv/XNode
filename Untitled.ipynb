```python
def search_manifest(timestamp):
    for row in range(len(manifest)):
        if manifest[row]['timestamp'] - timestamp > 0:
            return row
```

```python
def find_timestamp(n):
    return datetime.datetime(2000 + year[n][0],month[n][0],date[n][0],hour[n][0],minute[n][0],second[n][0])
    
def time_difference(dt1, dt2):
    seconds = (dt1 - dt2).total_seconds()
    return seconds / 60

def manifest_to_datetime(timestamp, ignore_times = False):
    n = timestamp
    m = n % 100
    n = int(n / 100)
    h = n % 100
    n = int(n / 100)
    d = n % 100
    n = int(n / 100)
    mo = n % 100
    n = int(n / 100)
    y = 2000 + n % 100
    return datetime.datetime(y, mo, d, h, m)

def find_bridge(n):
    if n <= 5:
        return '122.4'
    return '121.7'
```

```python
def set_entries(entries, old_accel):
    new_accel = []
    for entry in entries:
        new_accel.append(old_accel[entry])
    return new_accel
```

```python
#param threshold - minimum length of data in seconds
#return filter_length_results - dictionary of kept and discarded results
#       variance_stats - dictionary of variance statistics

def filter_length(accel, threshold):
    filter_length_results = {'kept':[],'discarded':[]}
    variance_data = {'kept':[],'discarded':[]}
    variance_stats = {}
#     if keep_entries != accel:
#         accel = set_entries(keep_entries, accel)
    for entry in range(len(accel)):  
        if len(accel[entry]['z']) >= threshold * 100:
            filter_length_results['kept'].append(entry)
            variance_data['kept'].append(np.var(accel[entry]['z']))
        else:
            filter_length_results['discarded'].append(entry)
            variance_data['discarded'].append(np.var(accel[entry]['z']))
    for category in list(variance_data.keys()):
        variance_stats[category] = {'min':np.min(variance_data[category]),'max':np.max(variance_data[category]),'median':np.median(variance_data[category])}
    return filter_length_results, variance_stats

#param threshold - minimum variance of raw data
#return filter_variance_results - dictionary of kept and discarded results

def filter_variance(accel, threshold):
    filter_variance_results = {'kept':[],'discarded':[]}
#     if keep_entries != accel:
#         accel = set_entries(keep_entries, accel)
    for entry in range(len(accel)):
        if np.var(accel[entry]['z']) >= threshold:
            filter_variance_results['kept'].append(entry)
        else:
            filter_variance_results['discarded'].append(entry)
    return filter_variance_results
```

```python
#param dates - list of dates to match in yymmdd format (e.g. [180921,180922])
#return matches - dictionary of matched dataset indices

def match_dates(dates):
    date_matches = {}
    for date in dates:
        date_matches[date] = []
        date_dt = manifest_to_datetime(date * 100**2)
        for entry in range(len(accel)):
            accel_dt = find_timestamp(entry)
            if (date_dt.strftime("%y") == accel_dt.strftime("%y")) and (date_dt.strftime("%m") == accel_dt.strftime("%m")) and (date_dt.strftime("%d") == accel_dt.strftime("%d")): 
                date_matches[date].append(entry)
    return date_matches

#param date_indices - entry of date_matches
#param timestamp - desired timestamp
#param north_window - tolerance for northbound train
#param south_window - tolerance for southbound train
#return matches - dictionary of matched dataset indices

def match_window(manifest_date, direction, north_window = 15, south_window = 60):
    window_matches = {'all':[],'121.7':[],'122.4':[]}
    date_dt = manifest_to_datetime(manifest_date)
    for n in range(len(accel)):
        accel_dt = find_timestamp(n)
        diff = time_difference(date_dt, accel_dt)
        bridge = find_bridge(nodeID[n][0])
        if direction == 'S' and diff >= 0 and diff <= south_window:
            window_matches[bridge].append(n)
            window_matches['all'].append(n)
        if direction == 'N' and diff <= 0 and abs(diff) <= north_window:
            window_matches[bridge].append(n)
            window_matches['all'].append(n)
    return window_matches
```

```python
# first iteration: filter on legth then variance

length_threshold = 10

print('initial total data entries: \t\t\t', len(accel))
filter_length_results, variance_stats = filter_length(accel, length_threshold)
new_accel = set_entries(filter_length_results['kept'],accel)
print('data entries after filtering by length: \t', len(new_accel))

median_kept = variance_stats['kept']['median']
print('median variance of kept entries: \t\t', median_kept)
median_discarded = variance_stats['discarded']['median']
print('median variance of discarded entries: \t\t', median_discarded)

variance_threshold = np.mean([median_kept, median_discarded])
print('average variance for threshold: \t\t', variance_threshold)

filter_variance_results = filter_variance(new_accel, variance_threshold)
new_accel = set_entries(filter_variance_results['kept'],new_accel)
print('data entries after filtering by variance: \t', len(new_accel))
```

```python
# second iteration: filter on variance then length

length_threshold = 10
variance_threshold = 200

print('initial total data entries: \t\t\t', len(accel))
filter_variance_results = filter_variance(accel, variance_threshold)
new_accel = set_entries(filter_variance_results['kept'],accel)
print('data entries after filtering by variance: \t', len(new_accel))

filter_length_results, variance_stats = filter_length(new_accel, length_threshold)
new_accel = set_entries(filter_length_results['kept'],new_accel)
print('data entries after filtering by length: \t', len(new_accel))
```
